```python
train_for_eda = pd.read_csv('train_for_EDA.csv')
oof = pd.read_csv('../../output/20210309204842_31_use_max_depth-1/oof/oof.csv')
```

```python
train_for_eda['pred'] = oof['likes']
train_for_eda['diff_act_pred'] = train_for_eda['likes'] - train_for_eda['pred']
train_for_eda['diff_abs_act_pred'] = abs(train_for_eda['likes'] - train_for_eda['pred'])
```

```python
def calc_score(true, pred):
    return (np.log(true+1) - np.log(pred+1))**2

train_for_eda['score'] = calc_score(train_for_eda['likes'], train_for_eda['pred'])
```

```python
train_for_eda.sort_values('diff_abs_act_pred',ascending=False)[['likes','pred','diff_act_pred','diff_abs_act_pred','score']].head(5)
```

```python
train_for_eda.sort_values('score',ascending=False)[['likes','pred','diff_act_pred','diff_abs_act_pred','score']].head(10)
```
